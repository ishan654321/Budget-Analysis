```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
!pip install squarify
import squarify
```

```python
data = pd.read_csv("/content/India_budget_2021.csv")
data.dropna()
```

```python
data = data.iloc[[0, 8, 11, 14, 18, 23, 41, 42, 43], :]
```

```python
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in â‚¹crores)': 592971.08}
data = data.append(row, ignore_index=True)
```

```python
data = data.sort_values(by='Fund allotted(in â‚¹crores)', ascending=False)
```

```python
data['Fund allotted(in â‚¹crores)'] = data['Fund allotted(in â‚¹crores)'].apply(lambda x: "{:,.2f}".format(x))
#apply(lambda x: "{:,.2f}".format(x))
#str.replace(',', '').astype(float)
```

```python
data_2023 = pd.read_csv("/content/indan_ budget_ 2023.csv")
data_2023.dropna()
```

```python
data_2023 = data_2023.iloc[[0, 8, 11, 14, 18, 23, 41, 42, 43], :]
```

```python
row_2023 = {'Department /Ministry': 'OTHERS', 'Fund allotted(in â‚¹crores)': 592971.08}
data_2023 = data_2023.append(row_2023, ignore_index=True)
```

```python
data_2023 = data_2023.sort_values(by='Fund allotted(in â‚¹crores)', ascending=False)
```

```python
data_2023['Fund allotted(in â‚¹crores)'] = data_2023['Fund allotted(in â‚¹crores)'].apply(lambda x: "{:,.2f}".format(x))
#apply(lambda x: "{:,.2f}".format(x))
#str.replace(',', '').astype(float)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Department /Ministry'], data['Fund allotted(in â‚¹crores)'], color='skyblue')
plt.xlabel('Department / Ministry')
plt.ylabel('Fund allotted (in â‚¹crores)')
plt.title('India Budget 2021 - Department-wise Allocation')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(data['Fund allotted(in â‚¹crores)'], labels=data['Department /Ministry'], autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow=True, colors=plt.cm.Paired(np.linspace(0, 1, len(data))))
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.title("Distribution of The Budget", fontsize=20)
plt.legend(data['Department /Ministry'], loc='upper right', bbox_to_anchor=(-0.1, 1.), fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(15, 12))
colors = plt.cm.Paired(np.linspace(0, 1, len(data)))
squarify.plot(sizes=data['Fund allotted(in â‚¹crores)'], label=data['Department /Ministry'], color=colors, alpha=0.7)
plt.axis('off')
plt.title('India Budget 2021 - Department-wise Allocation (TreeMap)')
plt.show()
```
